# AM Signals and Demodulation: DSB-LC, DSB-SC, Envelope, Coherent, and IQ

This notebook develops the mathematical theory behind amplitude-modulated passband signals and the three demodulation families covered by the workshop: envelope detection, coherent demodulation, and IQ demodulation.

We use angular frequency $\omega = 2\pi f$. The baseband signal is denoted by $m(t)$, the carrier angular frequency by $\omega_c$, and the carrier phase by $\theta_c$. Unless stated otherwise, $m(t)$ is real valued, low-pass, and band-limited to $|\omega| \leq W$, with $W \ll \omega_c$ so that the positive and negative sidebands around the carrier do not overlap.

For deterministic signals we use the Fourier transform pair

$$
X(\omega)=\int_{-\infty}^{\infty}x(t)e^{-j\omega t}\,dt,
\qquad
x(t)=\frac{1}{2\pi}\int_{-\infty}^{\infty}X(\omega)e^{j\omega t}\,d\omega.
$$

For random processes, $R_x(\tau)=\mathbb{E}\{x(t)x(t+\tau)\}$ denotes the autocorrelation of a WSS process and $S_x(\omega)$ denotes its two-sided PSD.


## 1. Low-Pass, Passband, and Complex-Envelope Descriptions

A real passband signal centered at $\omega_c$ can be represented through a complex envelope $u(t)$ as

$$
s(t)=\operatorname{Re}\{u(t)e^{j\omega_c t}\}.
$$

Writing $u(t)=I(t)+jQ(t)$ gives the quadrature form

$$
s(t)=I(t)\cos(\omega_c t)-Q(t)\sin(\omega_c t).
$$

Thus $I(t)$ is the in-phase component and $Q(t)$ is the quadrature component. The equivalent envelope and phase are

$$
A(t)=|u(t)|=\sqrt{I^2(t)+Q^2(t)},
\qquad
\phi(t)=\arg u(t)=\operatorname{atan2}(Q(t),I(t)).
$$

Then

$$
s(t)=A(t)\cos(\omega_c t+\phi(t)).
$$

This representation is exact for any real passband signal whose spectrum is concentrated near $\pm\omega_c$. Demodulation can therefore be interpreted as estimating $u(t)$ or some function of $u(t)$ from $s(t)$.


## 2. Double-Sideband Suppressed-Carrier AM (DSB-SC)

In DSB-SC modulation the carrier is not transmitted as an independent spectral line. The transmitted waveform is

$$
s_{\mathrm{SC}}(t)=A_c m(t)\cos(\omega_c t),
$$

where $A_c$ is a carrier-amplitude scale. Using

$$
\cos(\omega_c t)=\frac{1}{2}\left(e^{j\omega_c t}+e^{-j\omega_c t}\right),
$$

the spectrum is

$$
S_{\mathrm{SC}}(\omega)
=\frac{A_c}{2}\left[M(\omega-\omega_c)+M(\omega+\omega_c)\right].
$$

The baseband spectrum is copied to two sidebands centered at $+\omega_c$ and $-\omega_c$. If $M(\omega)$ is supported on $|\omega|\leq W$, then DSB-SC occupies approximately

$$
\omega_c-W \leq |\omega| \leq \omega_c+W,
$$

so its positive-frequency transmission bandwidth is $2W$.

The complex envelope of a phase-aligned DSB-SC signal is real:

$$
u_{\mathrm{SC}}(t)=A_c m(t).
$$

The sign of $m(t)$ is therefore encoded as a $\pi$ phase reversal of the passband carrier. This fact is crucial: an envelope magnitude detector sees $|m(t)|$, not $m(t)$.


## 3. DSB-SC as a Random Process

Assume $m(t)$ is zero-mean WSS with autocorrelation $R_m(\tau)$ and PSD $S_m(\omega)$. The modulated signal

$$
s_{\mathrm{SC}}(t)=A_c m(t)\cos(\omega_c t)
$$

is generally cyclostationary rather than WSS, because multiplication by a deterministic periodic carrier introduces dependence on the absolute time $t$.

Using the symmetric-lag convention,

$$
R_s(t,\tau)=\mathbb{E}\left\{s\left(t+\frac{\tau}{2}\right)s\left(t-\frac{\tau}{2}\right)\right\},
$$

we obtain

$$
\begin{aligned}
R_{s_{\mathrm{SC}}}(t,\tau)
&=A_c^2 R_m(\tau)
\cos\left(\omega_c\left(t+\frac{\tau}{2}\right)\right)
\cos\left(\omega_c\left(t-\frac{\tau}{2}\right)\right) \\
&=\frac{A_c^2}{2}R_m(\tau)
\left[\cos(\omega_c\tau)+\cos(2\omega_c t)\right].
\end{aligned}
$$

Averaging over one carrier period $T_c=2\pi/\omega_c$ removes the explicit $t$ dependence:

$$
\overline{R}_{s_{\mathrm{SC}}}(\tau)
=\frac{1}{T_c}\int_0^{T_c}R_{s_{\mathrm{SC}}}(t,\tau)\,dt
=\frac{A_c^2}{2}R_m(\tau)\cos(\omega_c\tau).
$$

The corresponding cycle-averaged PSD is

$$
\overline{S}_{s_{\mathrm{SC}}}(\omega)
=\frac{A_c^2}{4}\left[S_m(\omega-\omega_c)+S_m(\omega+\omega_c)\right].
$$

The average transmitted power is

$$
P_{\mathrm{SC}}=\overline{R}_{s_{\mathrm{SC}}}(0)=\frac{A_c^2}{2}R_m(0).
$$


## 4. Double-Sideband Large-Carrier AM (DSB-LC)

In conventional AM, a carrier bias is added before modulation:

$$
s_{\mathrm{LC}}(t)=A_c\left[1+k_a m(t)\right]\cos(\omega_c t),
$$

where $k_a$ is the amplitude sensitivity. The instantaneous amplitude factor is

$$
a(t)=A_c\left[1+k_a m(t)\right].
$$

The usual no-overmodulation condition is

$$
1+k_a m(t)\geq 0 \qquad \text{for all operating values of } m(t).
$$

If $|m(t)|\leq m_{\max}$, the modulation index is

$$
\mu=k_a m_{\max},
$$

and no overmodulation requires

$$
0\leq \mu \leq 1.
$$

The deterministic spectrum is

$$
S_{\mathrm{LC}}(\omega)
=\pi A_c\left[\delta(\omega-\omega_c)+\delta(\omega+\omega_c)\right]
+\frac{A_c k_a}{2}\left[M(\omega-\omega_c)+M(\omega+\omega_c)\right].
$$

Thus DSB-LC contains a discrete carrier at $\pm\omega_c$ plus the same pair of sidebands as DSB-SC. If $m(t)$ is zero mean with average power $P_m=\mathbb{E}\{m^2(t)\}$, the time-average power is

$$
P_{\mathrm{LC}}=\frac{A_c^2}{2}\left(1+k_a^2P_m\right).
$$

The carrier power is $A_c^2/2$ and the sideband power is $A_c^2k_a^2P_m/2$.


## 5. Envelope and Analytic Signal

The analytic signal associated with a real passband waveform $s(t)$ is

$$
s_a(t)=s(t)+j\mathcal{H}\{s(t)\},
$$

where $\mathcal{H}\{\cdot\}$ is the Hilbert transform. Its magnitude is the passband envelope:

$$
A_s(t)=|s_a(t)|.
$$

If

$$
s(t)=a(t)\cos(\omega_c t)
$$

and $a(t)$ is sufficiently low-pass compared with $\omega_c$, then the Bedrosian separation condition gives approximately

$$
s_a(t)\approx a(t)e^{j\omega_c t},
\qquad
A_s(t)\approx |a(t)|.
$$

This approximation is the mathematical basis of envelope detection. It also shows why the detector recovers different objects for DSB-LC and DSB-SC.


## 6. Envelope Detection for DSB-LC

For DSB-LC,

$$
s_{\mathrm{LC}}(t)=A_c\left[1+k_a m(t)\right]\cos(\omega_c t).
$$

If the no-overmodulation condition holds, then $1+k_a m(t)\geq 0$, so

$$
A_{\mathrm{LC}}(t)=A_c\left[1+k_a m(t)\right].
$$

An ideal envelope detector followed by DC removal gives

$$
A_{\mathrm{LC}}(t)-A_c=A_c k_a m(t),
$$

and therefore

$$
\widehat{m}(t)=\frac{A_{\mathrm{LC}}(t)-A_c}{A_c k_a}=m(t).
$$

Thus envelope detection is correct for conventional AM only under the operating condition that the transmitted amplitude never changes sign.

If overmodulation occurs, there exist times for which $1+k_a m(t)<0$. The envelope becomes

$$
A_{\mathrm{LC}}(t)=A_c|1+k_a m(t)|,
$$

which folds the negative parts upward. The mapping is nonlinear and no longer invertible from magnitude alone.


## 7. Envelope Detection for DSB-SC

For DSB-SC,

$$
s_{\mathrm{SC}}(t)=A_c m(t)\cos(\omega_c t).
$$

The low-pass factor multiplying the carrier is $a(t)=A_c m(t)$, which may be positive or negative. Hence the analytic-envelope magnitude is approximately

$$
A_{\mathrm{SC}}(t)=A_c|m(t)|.
$$

The sign of $m(t)$ is lost. Therefore an envelope detector cannot recover a general signed message from DSB-SC modulation.

If $m(t)$ is zero-mean Gaussian with variance $\sigma_m^2$, then $|m(t)|$ is half-normal at each fixed time. Its mean and variance are

$$
\mathbb{E}\{|m(t)|\}=\sigma_m\sqrt{\frac{2}{\pi}},
\qquad
\operatorname{var}(|m(t)|)=\sigma_m^2\left(1-\frac{2}{\pi}\right).
$$

Consequently, a DSB-SC envelope detector introduces a positive DC component and changes the second-order structure of the process. It is a magnitude tracker, not a signed coherent receiver.


## Bipolar Message Behavior in LC and SC Modulation

A bipolar message is a signal $m(t)$ that can take positive and negative values. This is the normal situation for audio waveforms, zero-mean random processes, and many baseband communication signals. The important modulation question is therefore: what does the passband waveform do when $m(t)$ crosses zero and changes polarity?

In DSB-LC, the transmitted signal is

$$
s_{\mathrm{LC}}(t)=A_c\left[1+k_a m(t)\right]\cos(\omega_c t).
$$

The carrier term $1$ shifts the bipolar message upward before modulation. If the no-overmodulation condition

$$
1+k_a m(t)\geq 0
$$

holds for all $t$, then the multiplier of the carrier never becomes negative. The passband signal keeps the same carrier phase, and the information appears as a slow expansion and contraction of the carrier envelope. In this case, positive values of $m(t)$ make the envelope larger than $A_c$, while negative values make it smaller than $A_c$ but still nonnegative. This is why a simple envelope detector can recover the bipolar message after subtracting the carrier level.

In DSB-SC, the transmitted signal is

$$
s_{\mathrm{SC}}(t)=A_c m(t)\cos(\omega_c t).
$$

Here there is no positive carrier offset. When $m(t)>0$, the passband waveform is in phase with the carrier; when $m(t)<0$, it is equivalent to

$$
A_c m(t)\cos(\omega_c t)=A_c\lvert m(t)\rvert\cos(\omega_c t+\pi).
$$

Thus a polarity change in the baseband message becomes a $\pi$ phase reversal in the passband waveform. The physical envelope magnitude is only $A_c\lvert m(t)\rvert$, so the envelope does not contain the sign of $m(t)$. The sign is carried by carrier phase, not by envelope height.

This is the central phenomenon: LC converts a bipolar message into a nonnegative amplitude variation around a transmitted carrier reference, while SC leaves the message bipolar and stores polarity as carrier phase. LC spends extra carrier power to make envelope detection possible; SC saves that carrier power but requires coherent or IQ demodulation to recover the sign of the original message.


## 8. Coherent Demodulation

Coherent demodulation multiplies the received passband signal by a synchronized local oscillator and then low-pass filters the product. Let

$$
r(t)=s(t)\cos(\omega_c t+\phi),
$$

where $\phi$ is the phase error between the receiver oscillator and the transmitter carrier.

For DSB-SC,

$$
\begin{aligned}
r_{\mathrm{SC}}(t)
&=A_c m(t)\cos(\omega_c t)\cos(\omega_c t+\phi) \\
&=\frac{A_c}{2}m(t)\cos\phi
+\frac{A_c}{2}m(t)\cos(2\omega_c t+\phi).
\end{aligned}
$$

After an ideal LPF that passes the baseband but rejects the $2\omega_c$ term,

$$
y_{\mathrm{SC}}(t)=\operatorname{LPF}\{2r_{\mathrm{SC}}(t)\}=A_c m(t)\cos\phi.
$$

The factor 2 is a conventional receiver gain used to compensate for the product-to-baseband factor $1/2$. If $\phi=0$, the receiver recovers $A_c m(t)$. If $\phi=\pi/2$, the output is zero. If $\phi=\pi$, the output is sign reversed.

For DSB-LC,

$$
y_{\mathrm{LC}}(t)=\operatorname{LPF}\{2s_{\mathrm{LC}}(t)\cos(\omega_c t+\phi)\}
=A_c\left[1+k_a m(t)\right]\cos\phi.
$$

After removing the DC component and rescaling, the message is recovered when $\cos\phi\neq 0$.


## 9. Coherent Demodulation in the Frequency Domain

For DSB-SC, multiplying by $2\cos(\omega_c t+\phi)$ produces two spectral translations:

$$
2s_{\mathrm{SC}}(t)\cos(\omega_c t+\phi)
=A_c m(t)\cos\phi + A_c m(t)\cos(2\omega_c t+\phi).
$$

The first term lies at baseband and the second term lies around $\pm 2\omega_c$. The low-pass filter must satisfy

$$
W < \omega_{\mathrm{LPF}} < 2\omega_c-W
$$

so that it passes the message spectrum and rejects the double-frequency image.

The baseband output PSD for zero-mean WSS $m(t)$ is therefore

$$
S_y(\omega)=A_c^2\cos^2\phi\;S_m(\omega).
$$

The output autocorrelation is

$$
R_y(\tau)=A_c^2\cos^2\phi\;R_m(\tau).
$$

Thus coherent demodulation preserves the complete second-order shape of the baseband process. Phase error changes only the recovered scale, except at the quadrature-null case where $\cos\phi=0$.


## 10. IQ Demodulation

IQ demodulation estimates both quadrature projections of the passband signal. Define

$$
I(t)=\operatorname{LPF}\{2s(t)\cos(\omega_c t)\},
$$

$$
Q(t)=\operatorname{LPF}\{-2s(t)\sin(\omega_c t)\}.
$$

If

$$
s(t)=\operatorname{Re}\{u(t)e^{j\omega_c t}\}
=I_u(t)\cos(\omega_c t)-Q_u(t)\sin(\omega_c t),
$$

then ideal IQ demodulation gives

$$
I(t)=I_u(t),
\qquad
Q(t)=Q_u(t),
$$

provided the LPF rejects the images around $2\omega_c$.

Equivalently, complex mixing gives

$$
\operatorname{LPF}\{2s(t)e^{-j\omega_c t}\}=u(t).
$$

This is the cleanest theoretical form: IQ demodulation recovers the complex envelope.


## 11. IQ View of DSB-SC and DSB-LC

For phase-aligned DSB-SC,

$$
s_{\mathrm{SC}}(t)=A_c m(t)\cos(\omega_c t)
=\operatorname{Re}\{A_c m(t)e^{j\omega_c t}\}.
$$

Thus

$$
u_{\mathrm{SC}}(t)=A_c m(t),
\qquad
I(t)=A_c m(t),
\qquad
Q(t)=0.
$$

If the carrier has an unknown constant phase $\theta$, then

$$
s_{\mathrm{SC}}(t)=A_c m(t)\cos(\omega_c t+\theta)
=\operatorname{Re}\{A_c m(t)e^{j\theta}e^{j\omega_c t}\}.
$$

The IQ output becomes

$$
I(t)=A_c m(t)\cos\theta,
\qquad
Q(t)=A_c m(t)\sin\theta.
$$

The message is recovered by projecting onto the correct carrier phase:

$$
\widehat{m}(t)=\frac{I(t)\cos\theta+Q(t)\sin\theta}{A_c}.
$$

The magnitude

$$
\sqrt{I^2(t)+Q^2(t)}=A_c|m(t)|
$$

again loses the sign, so IQ magnitude alone behaves like an envelope detector for DSB-SC. The sign is carried by the complex-envelope phase, which jumps by $\pi$ when $m(t)$ changes sign.

For DSB-LC under no overmodulation,

$$
u_{\mathrm{LC}}(t)=A_c\left[1+k_a m(t)\right],
$$

so $I(t)$ contains the carrier bias plus the scaled message and $Q(t)=0$ in the phase-aligned case.


## 12. Envelope, Coherent, and IQ Receivers Compared

| Modulation | Receiver | Ideal output | Main condition | Main failure mode |
|---|---|---|---|---|
| DSB-LC | Envelope | $A_c[1+k_a m(t)]$ | $1+k_a m(t)\geq 0$ | Overmodulation folds the waveform |
| DSB-LC | Coherent | $A_c[1+k_a m(t)]\cos\phi$ | Carrier frequency and phase known up to $\cos\phi\neq 0$ | Phase quadrature null |
| DSB-LC | IQ | $A_c[1+k_a m(t)]e^{j\theta}$ | LPF separates baseband from $2\omega_c$ image | Carrier/phase estimation error |
| DSB-SC | Envelope | $A_c\lvert m(t)\rvert$ | Useful only if the message is nonnegative or only magnitude is needed | Sign is lost |
| DSB-SC | Coherent | $A_c m(t)\cos\phi$ | Carrier phase synchronized | Phase quadrature null or scale error |
| DSB-SC | IQ | $A_c m(t)e^{j\theta}$ | Complex carrier reference available or estimable | Magnitude-only processing loses sign |

The essential distinction is that envelope detection is a magnitude operation, coherent demodulation is a signed projection onto a carrier reference, and IQ demodulation recovers the full complex envelope.

The difference between LC and SC is the transmitted carrier term. In DSB-LC, the signal is $A_c[1+k_a m(t)]\cos(\omega_c t)$, so the constant $1$ sends a strong carrier component even when $m(t)=0$. That carrier gives the waveform a positive amplitude reference when $1+k_a m(t)\geq 0$, which is why an envelope detector can read the message from the passband magnitude.

In DSB-SC, the signal is $A_c m(t)\cos(\omega_c t)$, so there is no independent carrier reference. When $m(t)$ changes sign, the passband signal changes phase by $\pi$ rather than producing a negative envelope. A magnitude-only detector cannot distinguish $m(t)$ from $-m(t)$, so SC saves carrier power but requires coherent phase information, usually through a synchronized oscillator or an IQ receiver.


## Staged Experiment Structure

The assigned experiment is organized in four stages:

I. **Input Signal**: define the rectangular pulse and its sampling convention.

II. **DSB Modulation**: generate DSB-LC and DSB-SC passband signals for the requested modulation indices.

III. **AWGN Channel**: add passband white Gaussian noise at the requested SNR values.

IV. **Demodulation**: recover the message using envelope, coherent, and IQ receivers, then compare waveform, ACF, PSD, and bandwidth behavior.


## I. Input Signal

The empirical study uses a finite rectangular pulse as the input signal. The number of samples is

$$
N=4096,
$$

and the requested sampling increment is

$$
\Delta t=\frac{1}{N}.
$$

Therefore the frequency-axis computations use the equivalent sampling rate

$$
f_s=\frac{1}{\Delta t}=N=4096.
$$

The raw pulse indicator has amplitude 1 and duration 2048 samples, but the actual input message is centered so that it has no DC component:

$$
m[n]=
\begin{cases}
\frac{1}{2}, & 0\leq n<2048,\\
-\frac{1}{2}, & 2048\leq n<4096.
\end{cases}
$$

Its discrete-time duty cycle is $1/2$, and the sample mean of $m[n]$ is exactly zero. More generally, for duty cycle $D=\tau/T$, the centered rectangular message has high level $1-D$ and low level $-D$. Centering removes the spectral line at zero frequency; the remaining spectrum contains only the AC transition content. Because the pulse has sharp edges, its spectrum has sinc-like sidelobes and non-negligible high-frequency content.

For the empirical implementation, this one-high/one-low record is treated as one period of a rectangular waveform. Therefore the input fundamental frequency is

$$
f_m=\frac{1}{N\Delta t},
$$

and the carrier is chosen as $f_c=30f_m$.


## II. DSB Modulation

For DSB-LC, the modulation index is imposed through

$$
s_{\mathrm{LC},\mu}[n]=A_c\left[1+\mu m[n]\right]\cos(2\pi f_c n\Delta t),
\qquad
\mu\in\{0.72,1.00,1.23\}.
$$

In the numerical experiment the carrier is not chosen independently: it is fixed at $f_c=30f_m$, where $f_m$ is the input fundamental frequency defined by the full record length.

Because the assigned zero-mean message satisfies $m[n]\in\{1/2,-1/2\}$, all three LC cases remain nonnegative in the envelope multiplier $1+\mu m[n]$. For the largest requested value, $\mu=1.23$, the minimum multiplier is $1-\mu/2=0.385>0$. Thus these values do not represent overmodulation; instead they represent increasing envelope depth and sideband power.

For DSB-SC, the transmitted signal is

$$
s_{\mathrm{SC}}[n]=A_c m[n]\cos(2\pi f_c n\Delta t).
$$

Since the assigned rectangular message is zero-mean and bipolar, the SC complex envelope is signed. A magnitude envelope detector returns $A_c|m[n]|$ in the noiseless ideal case, so it loses the sign unless coherent or IQ demodulation is used.


## III. AWGN Channel

The channel adds white Gaussian noise at passband. For a target SNR in dB,

$$
\sigma_n^2=\frac{P_s}{10^{\mathrm{SNR}_{\mathrm{dB}}/10}},
\qquad
r[n]=s[n]+n[n],
$$

where $P_s$ is the empirical passband signal power. The experiment uses

$$
\mathrm{SNR}_{\mathrm{dB}}\in\{30,10,3\}.
$$

The channel stage is evaluated both directly, through passband ACF/PSD/bandwidth estimates, and indirectly through the degradation it causes in each demodulator output.


## IV. Demodulation

The demodulators are evaluated as follows.

1. **Envelope detector.** A diode peak detector with RC discharge is used. The good-inertia case suppresses carrier ripple while still following the rectangular pulse edges. The bad-inertia case uses a discharge that is too slow, so the recovered envelope decays sluggishly after the falling edge.

2. **Coherent detector.** The receiver multiplies by $2\cos(2\pi f_c n\Delta t+\phi)$ and low-pass filters:
   $$
   \widehat{x}_{\phi}[n]=\operatorname{LPF}\{2r[n]\cos(2\pi f_c n\Delta t+\phi)\}.
   $$
   For SC, the ideal noiseless output is $A_c m[n]\cos\phi$. For LC, the ideal noiseless output is $A_c[1+\mu m[n]]\cos\phi$. The sweep uses
   $$
   \phi\in\left\{0,\frac{\pi}{16},\frac{2\pi}{16},\ldots,\frac{8\pi}{16}\right\}.
   $$

3. **IQ detector.** Complex demodulation estimates the complex envelope:
   $$
   \widehat{u}[n]=\operatorname{LPF}\{2r[n]e^{-j(2\pi f_c n\Delta t+\phi)}\}.
   $$
   A pure phase error rotates the complex envelope and can be corrected if the phase is known or estimated. IQ imbalance is modeled as unequal I/Q gains plus a small quadrature phase skew, so even after phase correction the complex-envelope geometry is distorted.

The empirical metrics report correlation, normalized mean-square error (NMSE), and recovered SNR against the rectangular message. A best affine fit is also reported for some comparisons to separate shape distortion from simple gain and DC errors.


## Quadrature and Canonical Representations for LC and SC

For every real passband signal centered at $f_c$, there are two equivalent descriptions.

The **quadrature representation** is

$$
s[n]=I[n]\cos(2\pi f_c n\Delta t)-Q[n]\sin(2\pi f_c n\Delta t),
$$

with complex envelope

$$
u[n]=I[n]+jQ[n].
$$

The **canonical envelope-phase representation** is

$$
s[n]=A[n]\cos(2\pi f_c n\Delta t+\varphi[n]),
$$

where

$$
A[n]=\sqrt{I^2[n]+Q^2[n]},
\qquad
\varphi[n]=\operatorname{atan2}(Q[n],I[n]).
$$

### DSB-LC

For the assigned LC signal,

$$
s_{\mathrm{LC},\mu}[n]=A_c\left[1+\mu m[n]\right]\cos(2\pi f_c n\Delta t).
$$

Thus its ideal quadrature components are

$$
I_{\mathrm{LC}}[n]=A_c\left[1+\mu m[n]\right],
\qquad
Q_{\mathrm{LC}}[n]=0.
$$

Since $m[n]\in\{1/2,-1/2\}$ and the requested $\mu$ values keep $1+\mu m[n]>0$, $I_{\mathrm{LC}}[n]>0$. Therefore the canonical representation is

$$
A_{\mathrm{LC}}[n]=A_c\left[1+\mu m[n]\right],
\qquad
\varphi_{\mathrm{LC}}[n]=0.
$$

So, for this zero-mean rectangular input, LC stores the information as a strictly positive envelope variation around a transmitted carrier reference.

### DSB-SC

For the assigned SC signal,

$$
s_{\mathrm{SC}}[n]=A_c m[n]\cos(2\pi f_c n\Delta t).
$$

The ideal quadrature components are

$$
I_{\mathrm{SC}}[n]=A_c m[n],
\qquad
Q_{\mathrm{SC}}[n]=0.
$$

Because this particular $m[n]$ is bipolar, the signed quadrature component must be converted to magnitude and phase in the canonical representation:

$$
A_{\mathrm{SC}}[n]=A_c |m[n]|,
\qquad
\varphi_{\mathrm{SC}}[n]=\begin{cases}0,&m[n]>0,\\ \pi,&m[n]<0.\end{cases}
$$

Thus negative values of $m[n]$ appear as the same positive magnitude with a phase jump of $\pi$.

Empirically, the IQ demodulator estimates $\widehat{u}[n]=\widehat{I}[n]+j\widehat{Q}[n]$. From that estimate we compute both

$$
(\widehat{I}[n],\widehat{Q}[n])
$$

and

$$
\widehat{A}[n]=|\widehat{u}[n]|,
\qquad
\widehat{\varphi}[n]=\operatorname{unwrap}(\arg \widehat{u}[n]).
$$

The numerical comparison below checks these two descriptions for LC and SC under both ideal IQ and imbalanced IQ conditions.


```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(20260428)

# Experiment constants
num_samples = 4096
rect_duration_samples = 2048
rect_amplitude = 1.0
sample_interval = 1.0 / num_samples
fs = 1.0 / sample_interval
T = num_samples * sample_interval
tau = rect_duration_samples * sample_interval
input_signal_frequency_hz = 1.0 / T
omega_0 = 2.0 * np.pi * input_signal_frequency_hz
input_tau_bandwidth_rad_s = 2.0 * np.pi / tau
fc = 30.0 * input_signal_frequency_hz
Ac = 1.0
input_harmonic_cases = {
    "1": np.array([1], dtype=int),
    "1, 3": np.array([1, 3], dtype=int),
    "1, 3, 5": np.array([1, 3, 5], dtype=int),
    "1, 3, 5, 7": np.array([1, 3, 5, 7], dtype=int),
    "1, 3, 5, 7, 9": np.array([1, 3, 5, 7, 9], dtype=int),
}
input_harmonic_bins = np.array([1, 3, 5], dtype=int)
input_harmonic_bandwidth_hz = float(input_harmonic_bins[-1] * input_signal_frequency_hz)
input_harmonic_bandwidth_rad_s = float(input_harmonic_bins[-1] * omega_0)
modulation_indices = [0.72, 1.00, 1.23]
snr_db_values = [30.0, 10.0, 3.0]
phi_values = np.arange(0.0, np.pi / 2.0 + 1e-12, np.pi / 16.0)
message_bandwidth_hz = 3.0 * input_signal_frequency_hz

# Envelope detector RC choices, represented by discharge cutoff f_rc = 1/(2*pi*RC).
# Good: slow enough to smooth carrier ripple but fast enough for the 2048-sample pulse.
# Bad: too slow after the falling edge.
envelope_rc_cases = {
    "good inertia": fs / 512.0,
    "bad inertia": fs / 4096.0,
}

iq_imbalance_cases = {
    "ideal IQ": {"gain": 0.0, "phase_deg": 0.0},
    "imbalanced IQ": {"gain": 0.08, "phase_deg": 5.0},
}

t = np.arange(num_samples) * sample_interval
carrier = np.cos(2.0 * np.pi * fc * t)


def rectangular_pulse(num_samples, duration_samples, amplitude=1.0, zero_mean=True):
    if not 0 < duration_samples <= num_samples:
        raise ValueError("duration_samples must satisfy 0 < duration_samples <= num_samples.")
    message = np.zeros(num_samples, dtype=float)
    message[:duration_samples] = amplitude
    if zero_mean:
        message = message - amplitude * duration_samples / num_samples
    return message


def fft_lowpass(x, sample_rate_hz, cutoff_hz, order=10):
    x = np.asarray(x)
    freqs = np.fft.rfftfreq(x.size, d=1.0 / sample_rate_hz)
    response = 1.0 / np.sqrt(1.0 + (freqs / cutoff_hz) ** (2 * order))
    if np.iscomplexobj(x):
        real_part = np.fft.irfft(np.fft.rfft(np.real(x)) * response, n=x.size)
        imag_part = np.fft.irfft(np.fft.rfft(np.imag(x)) * response, n=x.size)
        return real_part + 1j * imag_part
    return np.fft.irfft(np.fft.rfft(x) * response, n=x.size)


def diode_peak_envelope(signal, sample_rate_hz, discharge_cutoff_hz):
    rectified = np.maximum(signal, 0.0)
    rc_seconds = 1.0 / (2.0 * np.pi * discharge_cutoff_hz)
    discharge_alpha = np.exp(-1.0 / (sample_rate_hz * rc_seconds))
    envelope = np.empty_like(rectified)
    envelope[0] = rectified[0]
    for idx in range(1, rectified.size):
        discharged = discharge_alpha * envelope[idx - 1]
        envelope[idx] = rectified[idx] if rectified[idx] > discharged else discharged
    return envelope


def add_awgn(signal, snr_db, generator):
    power = float(np.mean(signal**2))
    noise_power = power / (10.0 ** (snr_db / 10.0))
    noise = generator.normal(scale=np.sqrt(noise_power), size=signal.size)
    return signal + noise


def best_affine(reference, estimate):
    reference = np.asarray(reference, dtype=float)
    estimate = np.asarray(estimate, dtype=float)
    design = np.column_stack([estimate, np.ones_like(estimate)])
    scale, offset = np.linalg.lstsq(design, reference, rcond=None)[0]
    fitted = scale * estimate + offset
    return fitted, float(scale), float(offset)


def metrics(reference, estimate, allow_affine=False):
    reference = np.asarray(reference, dtype=float)
    estimate = np.asarray(estimate, dtype=float)
    if allow_affine:
        estimate, scale, offset = best_affine(reference, estimate)
    else:
        scale, offset = 1.0, 0.0
    error = estimate - reference
    ref_power = max(float(np.mean(reference**2)), 1e-15)
    err_power = max(float(np.mean(error**2)), 1e-15)
    corr = float(np.corrcoef(reference, estimate)[0, 1]) if np.std(estimate) > 1e-12 else 0.0
    return {
        "corr": corr,
        "nmse": err_power / ref_power,
        "snr_db": 10.0 * np.log10(ref_power / err_power),
        "affine_scale": scale,
        "affine_offset": offset,
    }


def modulate(message, mode, mu):
    if mode == "LC":
        return Ac * (1.0 + mu * message) * carrier
    if mode == "SC":
        return Ac * message * carrier
    raise ValueError(f"Unknown modulation mode: {mode}")


def recover_lc_from_baseband(baseband, mu):
    return (baseband / Ac - 1.0) / mu


def recover_sc_from_baseband(baseband):
    return baseband / Ac


def envelope_demodulate(received, mode, mu, rc_cutoff_hz):
    envelope = diode_peak_envelope(received, fs, rc_cutoff_hz)
    if mode == "LC":
        recovered = (envelope / Ac - 1.0) / mu
    else:
        recovered = envelope / Ac
    return recovered, envelope


def coherent_demodulate(received, mode, mu, phi):
    mixed = 2.0 * received * np.cos(2.0 * np.pi * fc * t + phi)
    baseband = fft_lowpass(mixed, fs, cutoff_hz=1.35 * message_bandwidth_hz)
    if mode == "LC":
        return recover_lc_from_baseband(baseband, mu), baseband
    return recover_sc_from_baseband(baseband), baseband


def apply_iq_imbalance(complex_baseband, gain_imbalance, phase_skew_rad):
    i_part = np.real(complex_baseband)
    q_part = np.imag(complex_baseband)
    i_out = (1.0 + gain_imbalance) * i_part
    q_out = (1.0 - gain_imbalance) * (np.cos(phase_skew_rad) * q_part + np.sin(phase_skew_rad) * i_part)
    return i_out + 1j * q_out


def iq_demodulate(received, mode, mu, phi, gain_imbalance=0.0, phase_skew_rad=0.0, correct_phase=True):
    mixed = 2.0 * received * np.exp(-1j * (2.0 * np.pi * fc * t + phi))
    complex_baseband = fft_lowpass(mixed, fs, cutoff_hz=1.35 * message_bandwidth_hz)
    complex_baseband = apply_iq_imbalance(complex_baseband, gain_imbalance, phase_skew_rad)
    if correct_phase:
        complex_baseband = complex_baseband * np.exp(1j * phi)
    recovered_baseband = np.real(complex_baseband)
    if mode == "LC":
        return recover_lc_from_baseband(recovered_baseband, mu), complex_baseband
    return recover_sc_from_baseband(recovered_baseband), complex_baseband


def format_table(rows, columns, max_rows=None):
    if max_rows is not None:
        rows = rows[:max_rows]
    widths = []
    for column in columns:
        widths.append(max(len(column), max(len(str(row[column])) for row in rows) if rows else len(column)))
    header = " | ".join(column.ljust(width) for column, width in zip(columns, widths))
    sep = "-+-".join("-" * width for width in widths)
    body = [" | ".join(str(row[column]).ljust(width) for column, width in zip(columns, widths)) for row in rows]
    return "\n".join([header, sep, *body])


message = rectangular_pulse(num_samples, rect_duration_samples, rect_amplitude)
analysis_slice = slice(0, None)
message_eval = message[analysis_slice]

print(f"Number of samples: {num_samples}")
print(f"Rectangular pulse duration: {rect_duration_samples} samples")
print(f"Sample interval requested: {sample_interval:.6e}")
print(f"Equivalent sampling rate used for frequency axes: {fs:.1f} Hz")
print(f"Input fundamental frequency f_m: {input_signal_frequency_hz:.1f} Hz")
print(f"Input fundamental angular frequency omega_0: {omega_0:.6f} rad/s")
print(f"Rectangular pulse duration tau: {tau:.6f} s")
print(f"Tau-based one-sided bandwidth: {input_tau_bandwidth_rad_s / omega_0:.1f} omega_0")
print(f"Input harmonic truncation: {input_harmonic_bins.tolist()} odds, bandwidth = {input_harmonic_bins[-1]} omega_0")
print(f"Message sample mean: {np.mean(message):+.6f}")
print(f"Message levels: min={np.min(message):+.3f}, max={np.max(message):+.3f}")
print(f"Message peak amplitude: {np.max(np.abs(message)):.3f}")
print(f"Carrier f_c: {fc:.1f} Hz = {fc / input_signal_frequency_hz:.1f} f_m")
print(f"Baseband analysis cutoff: {message_bandwidth_hz:.1f} Hz")
```

```python
def ideal_complex_envelope(mode, mu):
    if mode == "LC":
        return Ac * (1.0 + mu * message).astype(complex)
    if mode == "SC":
        return Ac * message.astype(complex)
    raise ValueError(f"Unknown mode: {mode}")


def representation_summary(mode, mu, snr_db, phi, iq_case_name):
    transmitted = modulate(message, mode, mu)
    received = add_awgn(transmitted, snr_db, rng)
    iq_params = iq_imbalance_cases[iq_case_name]
    _, estimated_u = iq_demodulate(
        received,
        mode,
        mu,
        phi,
        gain_imbalance=iq_params["gain"],
        phase_skew_rad=np.deg2rad(iq_params["phase_deg"]),
        correct_phase=True,
    )
    reference_u = ideal_complex_envelope(mode, mu)
    reference_i = np.real(reference_u)
    reference_q = np.imag(reference_u)
    estimated_i = np.real(estimated_u)
    estimated_q = np.imag(estimated_u)
    reference_a = np.abs(reference_u)
    estimated_a = np.abs(estimated_u)
    active = reference_a > 0.05 * max(np.max(reference_a), 1e-12)
    reference_phase = np.unwrap(np.angle(reference_u[active])) if np.any(active) else np.array([0.0])
    estimated_phase = np.unwrap(np.angle(estimated_u[active])) if np.any(active) else np.array([0.0])
    phase_error = estimated_phase - reference_phase
    phase_error = np.angle(np.exp(1j * phase_error))
    i_metrics = metrics(reference_i[analysis_slice], estimated_i[analysis_slice], allow_affine=True)
    a_metrics = metrics(reference_a[analysis_slice], estimated_a[analysis_slice], allow_affine=True)
    q_rms = float(np.sqrt(np.mean(estimated_q[analysis_slice] ** 2)))
    i_rms = float(np.sqrt(np.mean(estimated_i[analysis_slice] ** 2)))
    return {
        "mode": mode,
        "mu": f"{mu:.2f}",
        "snr": f"{snr_db:.0f}",
        "iq_case": iq_case_name,
        "I_corr": f"{i_metrics['corr']:.3f}",
        "I_nmse": f"{i_metrics['nmse']:.3f}",
        "A_corr": f"{a_metrics['corr']:.3f}",
        "A_nmse": f"{a_metrics['nmse']:.3f}",
        "Q_over_I_rms": f"{q_rms / max(i_rms, 1e-12):.3f}",
        "phase_rms_rad": f"{float(np.sqrt(np.mean(phase_error**2))):.3f}",
        "reference_u": reference_u,
        "estimated_u": estimated_u,
    }


representation_mu = 1.00
representation_snr_db = 10.0
representation_phi = np.pi / 4.0
representation_rows = []
representation_data = {}
for mode in ["LC", "SC"]:
    for iq_case_name in ["ideal IQ", "imbalanced IQ"]:
        summary = representation_summary(mode, representation_mu, representation_snr_db, representation_phi, iq_case_name)
        representation_rows.append({key: value for key, value in summary.items() if not key.endswith("_u")})
        representation_data[(mode, iq_case_name)] = summary

print("Quadrature and canonical representation check, using IQ phase correction")
print(format_table(
    representation_rows,
    ["mode", "mu", "snr", "iq_case", "I_corr", "I_nmse", "A_corr", "A_nmse", "Q_over_I_rms", "phase_rms_rad"],
))

plot_view = slice(0, num_samples)
fig, axes = plt.subplots(4, 2, figsize=(15, 12), sharex=True)
for col, mode in enumerate(["LC", "SC"]):
    ideal = representation_data[(mode, "ideal IQ")]
    imbalanced = representation_data[(mode, "imbalanced IQ")]
    reference_u = ideal["reference_u"]
    axes[0, col].plot(t[plot_view], np.real(reference_u[plot_view]), color="black", lw=2.0, label="ideal I")
    axes[0, col].plot(t[plot_view], np.real(ideal["estimated_u"][plot_view]), lw=1.2, label="estimated I, ideal IQ")
    axes[0, col].plot(t[plot_view], np.real(imbalanced["estimated_u"][plot_view]), lw=1.2, ls="--", label="estimated I, imbalanced IQ")
    axes[0, col].set_title(f"{mode}: in-phase component")
    axes[0, col].set_ylabel("I[n]")
    axes[0, col].legend(fontsize=8)
    axes[0, col].grid(alpha=0.3)

    axes[1, col].plot(t[plot_view], np.imag(ideal["estimated_u"][plot_view]), lw=1.2, label="Q, ideal IQ")
    axes[1, col].plot(t[plot_view], np.imag(imbalanced["estimated_u"][plot_view]), lw=1.2, ls="--", label="Q, imbalanced IQ")
    axes[1, col].axhline(0.0, color="black", lw=0.8, alpha=0.6)
    axes[1, col].set_title(f"{mode}: quadrature leakage")
    axes[1, col].set_ylabel("Q[n]")
    axes[1, col].legend(fontsize=8)
    axes[1, col].grid(alpha=0.3)

    axes[2, col].plot(t[plot_view], np.abs(reference_u[plot_view]), color="black", lw=2.0, label="ideal A")
    axes[2, col].plot(t[plot_view], np.abs(ideal["estimated_u"][plot_view]), lw=1.2, label="estimated A, ideal IQ")
    axes[2, col].plot(t[plot_view], np.abs(imbalanced["estimated_u"][plot_view]), lw=1.2, ls="--", label="estimated A, imbalanced IQ")
    axes[2, col].set_title(f"{mode}: canonical envelope")
    axes[2, col].set_ylabel("A[n]")
    axes[2, col].legend(fontsize=8)
    axes[2, col].grid(alpha=0.3)

    active = np.abs(reference_u) > 0.05 * max(np.max(np.abs(reference_u)), 1e-12)
    phase_ideal = np.full_like(t, np.nan, dtype=float)
    phase_imbalanced = np.full_like(t, np.nan, dtype=float)
    phase_ideal[active] = np.angle(ideal["estimated_u"][active])
    phase_imbalanced[active] = np.angle(imbalanced["estimated_u"][active])
    axes[3, col].plot(t[plot_view], phase_ideal[plot_view], lw=1.2, label="phase, ideal IQ")
    axes[3, col].plot(t[plot_view], phase_imbalanced[plot_view], lw=1.2, ls="--", label="phase, imbalanced IQ")
    axes[3, col].axhline(0.0, color="black", lw=0.8, alpha=0.6)
    axes[3, col].set_title(f"{mode}: canonical phase on active samples")
    axes[3, col].set_xlabel("time [s]")
    axes[3, col].set_ylabel(r"$\varphi[n]$ [rad]")
    axes[3, col].legend(fontsize=8)
    axes[3, col].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Baseline figures: message and representative LC/SC waveforms.
example_mu = 1.00
example_snr_db = 30.0
example_lc = add_awgn(modulate(message, "LC", example_mu), example_snr_db, rng)
example_sc = add_awgn(modulate(message, "SC", example_mu), example_snr_db, rng)

fig, axes = plt.subplots(3, 1, figsize=(13, 8), sharex=False)
view = slice(0, num_samples)
axes[0].plot(t[view], message[view], lw=2.0)
axes[0].set_title("Zero-mean rectangular input: duration 2048 samples")
axes[0].set_xlabel("time [s]")
axes[0].set_ylabel("m(t)")
axes[0].grid(alpha=0.3)

axes[1].plot(t[view], example_lc[view], lw=1.0, label="DSB-LC, mu=1, SNR=30 dB")
axes[1].plot(t[view], 1.0 + example_mu * message[view], lw=1.5, ls="--", label="LC ideal upper envelope")
axes[1].plot(t[view], -(1.0 + example_mu * message[view]), lw=1.5, ls="--", color="tab:orange", label="LC ideal lower envelope")
axes[1].set_title("Representative LC passband waveform")
axes[1].set_xlabel("time [s]")
axes[1].set_ylabel("amplitude")
axes[1].legend(loc="upper right")
axes[1].grid(alpha=0.3)

axes[2].plot(t[view], example_sc[view], lw=1.0, label="DSB-SC, SNR=30 dB")
axes[2].plot(t[view], np.abs(message[view]), lw=1.5, ls="--", label="SC magnitude envelope")
axes[2].plot(t[view], -np.abs(message[view]), lw=1.5, ls="--", color="tab:orange")
axes[2].set_title("Representative SC passband waveform: message polarity is carrier phase")
axes[2].set_xlabel("time [s]")
axes[2].set_ylabel("amplitude")
axes[2].legend(loc="upper right")
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()


def rectangular_dft_magnitude_theory(harmonic_bins, record_length, pulse_length, zero_mean=True):
    harmonic_bins = np.asarray(harmonic_bins, dtype=int)
    magnitude = np.empty(harmonic_bins.size, dtype=float)
    zero = harmonic_bins == 0
    magnitude[zero] = 0.0 if zero_mean else pulse_length / record_length
    nonzero_bins = harmonic_bins[~zero]
    numerator = np.sin(np.pi * nonzero_bins * pulse_length / record_length)
    denominator = np.sin(np.pi * nonzero_bins / record_length)
    magnitude[~zero] = np.abs(numerator / denominator) / record_length
    return magnitude


def rectangular_pair_spectrum_case(tau_seconds, T_seconds, record_length, harmonic_limit):
    sample_interval_case = T_seconds / record_length
    pulse_length = int(round(tau_seconds / sample_interval_case))
    if not 0 < pulse_length <= record_length:
        raise ValueError("tau_seconds must satisfy 0 < tau_seconds <= T_seconds for the chosen record_length.")
    time_case = np.arange(record_length) * sample_interval_case
    case_message = rectangular_pulse(record_length, pulse_length, rect_amplitude)
    case_spectrum = np.fft.fft(case_message)
    harmonic_bins = np.arange(-harmonic_limit, harmonic_limit + 1, dtype=int)
    wrapped_bins = np.mod(harmonic_bins, record_length)
    calculated_magnitude = np.abs(case_spectrum[wrapped_bins]) / record_length
    theoretical_magnitude = rectangular_dft_magnitude_theory(np.abs(harmonic_bins), record_length, pulse_length)
    dc_idx = int(np.flatnonzero(harmonic_bins == 0)[0])
    omega0_case = 2.0 * np.pi / T_seconds
    return {
        "tau": tau_seconds,
        "T": T_seconds,
        "duty_cycle": tau_seconds / T_seconds,
        "pulse_length": pulse_length,
        "time": time_case,
        "message": case_message,
        "mean": float(np.mean(case_message)),
        "dc_magnitude": float(calculated_magnitude[dc_idx]),
        "harmonic_bins": harmonic_bins,
        "calculated_magnitude": calculated_magnitude,
        "theoretical_magnitude": theoretical_magnitude,
        "B_tau_over_omega0": T_seconds / tau_seconds,
        "B_T_over_omega0": float(input_harmonic_bins[-1]),
        "omega0": omega0_case,
    }


tau_T_cases = [
    (tau, T),
    (tau / 2.0, T),
    (tau / 4.0, T),
    (tau, 2.0 * T),
    (tau / 2.0, 2.0 * T),
]
pair_spectrum_cases = [rectangular_pair_spectrum_case(tau_case, T_case, num_samples, harmonic_limit=11) for tau_case, T_case in tau_T_cases]
pair_rows = [
    {
        "tau": f"{case['tau']:.3f}",
        "T": f"{case['T']:.3f}",
        "D=tau/T": f"{case['duty_cycle']:.3f}",
        "L": case["pulse_length"],
        "mean": f"{case['mean']:+.2e}",
        "DC mag": f"{case['dc_magnitude']:.2e}",
        "B_tau/omega0": f"{case['B_tau_over_omega0']:.1f}",
        "B_T/omega0": f"{case['B_T_over_omega0']:.1f}",
    }
    for case in pair_spectrum_cases
]
print("\nBilateral spectrum cases by tau and T")
print(format_table(pair_rows, ["tau", "T", "D=tau/T", "L", "mean", "DC mag", "B_tau/omega0", "B_T/omega0"]))

fig, axes = plt.subplots(len(pair_spectrum_cases), 3, figsize=(18, 2.8 * len(pair_spectrum_cases)))
for row, case in enumerate(pair_spectrum_cases):
    time_ax = axes[row, 0]
    spectrum_ax = axes[row, 1]
    bandwidth_ax = axes[row, 2]
    bins = case["harmonic_bins"]
    time_ax.step(case["time"], case["message"], where="post", lw=1.8, color="tab:blue")
    time_ax.axvline(case["tau"], color="tab:purple", lw=1.2, ls="--")
    time_ax.axvline(case["T"], color="0.25", lw=1.0, ls=":")
    time_ax.set_title(rf"$\tau={case['tau']:.3f}$ s, $T={case['T']:.3f}$ s, $D={case['duty_cycle']:.3f}$")
    time_ax.set_xlim(0.0, case["T"])
    level_margin = 0.1 * max(1.0, float(np.max(case["message"]) - np.min(case["message"])))
    time_ax.set_ylim(float(np.min(case["message"]) - level_margin), float(np.max(case["message"]) + level_margin))
    time_ax.set_ylabel("m(t)")
    time_ax.grid(alpha=0.3)

    spectrum_ax.vlines(bins, 0.0, case["calculated_magnitude"], lw=1.6, color="tab:blue", label="calculated" if row == 0 else None)
    spectrum_ax.scatter(bins, case["calculated_magnitude"], s=28, color="tab:blue", zorder=3)
    spectrum_ax.plot(bins, case["theoretical_magnitude"], marker="x", ms=6, lw=0.0, color="tab:orange", label="theoretical" if row == 0 else None)
    spectrum_ax.axvline(0.0, color="0.2", lw=0.8)
    spectrum_ax.set_title("bilateral spectrum")
    spectrum_ax.set_ylabel("magnitude")
    spectrum_ax.grid(alpha=0.3)
    if row == 0:
        spectrum_ax.legend(loc="upper right")

    b_tau = case["B_tau_over_omega0"]
    b_t_grid = case["B_T_over_omega0"]
    bandwidth_ax.hlines(0.65, -b_tau, b_tau, color="tab:purple", lw=3.0)
    bandwidth_ax.plot([-b_tau, b_tau], [0.65, 0.65], marker="|", color="tab:purple", ms=12, lw=0.0)
    bandwidth_ax.text(b_tau + 0.25, 0.65, rf"$B_{{\omega,\tau}}=\pm {b_tau:.1f}\omega_0$", va="center", color="tab:purple")
    bandwidth_ax.hlines(0.25, -b_t_grid, b_t_grid, color="tab:green", lw=3.0)
    bandwidth_ax.plot([-b_t_grid, b_t_grid], [0.25, 0.25], marker="|", color="tab:green", ms=12, lw=0.0)
    bandwidth_ax.text(b_t_grid + 0.25, 0.25, rf"retained grid $=\pm {b_t_grid:.1f}\omega_0$", va="center", color="tab:green")
    bandwidth_ax.axvline(0.0, color="0.2", lw=0.8)
    limit = max(11.5, b_tau + 1.5, b_t_grid + 1.5)
    bandwidth_ax.set_xlim(-limit, limit)
    bandwidth_ax.set_ylim(0.0, 0.9)
    bandwidth_ax.set_yticks([])
    bandwidth_ax.grid(axis="x", alpha=0.3)

axes[-1, 0].set_xlabel("time [s]")
axes[-1, 1].set_xlabel(r"harmonic index $k=\omega/\omega_0$")
axes[-1, 2].set_xlabel(r"angular frequency / $\omega_0$")
plt.tight_layout()
plt.show()

freq = np.fft.rfftfreq(message.size, d=1.0 / fs)
msg_spectrum = np.abs(np.fft.rfft(message)) / message.size
msg_spectral_energy = np.abs(np.fft.rfft(message)) ** 2 / message.size**2
msg_spectral_energy[1:-1] *= 2.0


harmonic_freq = input_harmonic_bins * input_signal_frequency_hz
harmonic_magnitude = msg_spectrum[input_harmonic_bins]
comparison_harmonic_bins = input_harmonic_cases["1, 3, 5, 7, 9"]
comparison_harmonic_freq = comparison_harmonic_bins * input_signal_frequency_hz
comparison_calculated_magnitude = msg_spectrum[comparison_harmonic_bins]
comparison_theoretical_magnitude = rectangular_dft_magnitude_theory(
    comparison_harmonic_bins,
    record_length=num_samples,
    pulse_length=rect_duration_samples,
)
spectrum_comparison_rows = [
    {
        "k": int(k),
        "calc_mag": f"{calc:.6f}",
        "theory_mag": f"{theory:.6f}",
        "abs_err": f"{abs(calc - theory):.2e}",
    }
    for k, calc, theory in zip(comparison_harmonic_bins, comparison_calculated_magnitude, comparison_theoretical_magnitude)
]
signed_harmonic_bins = np.concatenate([-comparison_harmonic_bins[::-1], comparison_harmonic_bins])
signed_omega_over_omega0 = signed_harmonic_bins.astype(float)
signed_calculated_magnitude = msg_spectrum[np.abs(signed_harmonic_bins)]
signed_theoretical_magnitude = rectangular_dft_magnitude_theory(
    np.abs(signed_harmonic_bins),
    record_length=num_samples,
    pulse_length=rect_duration_samples,
)
active_omega_over_omega0 = np.concatenate([-input_harmonic_bins[::-1], input_harmonic_bins]).astype(float)
input_harmonic_bandwidth_rows = []
for case_name, bins in input_harmonic_cases.items():
    bandwidth_hz = float(bins[-1] * input_signal_frequency_hz)
    bandwidth_rad_s = float(bins[-1] * omega_0)
    selected_energy = float(np.sum(msg_spectral_energy[bins]))
    input_harmonic_bandwidth_rows.append(
        {
            "harmonics": case_name,
            "band_hz": f"{bandwidth_hz:.1f}",
            "band_over_omega0": f"{bandwidth_rad_s / omega_0:.1f}",
            "band_rad_s": f"{bandwidth_rad_s:.3f}",
            "selected_energy": f"{selected_energy:.6f}",
        }
    )

print("Input deterministic bandwidth by retained odd-harmonic case")
print(format_table(input_harmonic_bandwidth_rows, ["harmonics", "band_hz", "band_over_omega0", "band_rad_s", "selected_energy"]))
print("\nInput calculated versus theoretical DFT magnitude")
print(format_table(spectrum_comparison_rows, ["k", "calc_mag", "theory_mag", "abs_err"]))

fig, axes = plt.subplots(
    2,
    1,
    figsize=(14, 7),
    sharex=True,
    gridspec_kw={"height_ratios": [3.0, 1.4]},
)
axes[0].vlines(signed_omega_over_omega0, 0.0, signed_calculated_magnitude, lw=2.0, color="tab:blue", label="calculated DFT")
axes[0].scatter(signed_omega_over_omega0, signed_calculated_magnitude, s=55, color="tab:blue", zorder=3)
axes[0].plot(signed_omega_over_omega0, signed_theoretical_magnitude, marker="x", ms=9, lw=0.0, color="tab:orange", label="theoretical")
for active_omega in active_omega_over_omega0:
    axes[0].axvline(active_omega, color="0.75", lw=0.8, ls=":", zorder=0)
axes[0].set_title("Rectangular pulse DFT magnitude: bilateral calculated vs theoretical")
axes[0].set_ylabel("magnitude")
axes[0].grid(alpha=0.3)
axes[0].legend()

case_labels = list(input_harmonic_cases.keys())
bandwidth_over_omega0 = [float(bins[-1]) for bins in input_harmonic_cases.values()]
bandwidth_marker_rows = [(r"$B_{\omega,\tau}=2\pi/\tau$", input_tau_bandwidth_rad_s / omega_0, "tab:purple", 2.4)]
for idx, (case_name, bins) in enumerate(input_harmonic_cases.items()):
    label = f"T grid {case_name}"
    color = "tab:green" if np.array_equal(bins, input_harmonic_bins) else "0.55"
    bandwidth_marker_rows.append((label, float(bins[-1]), color, 1.3 - 0.35 * idx))

for label, half_width, color, y_pos in bandwidth_marker_rows:
    axes[1].hlines(y_pos, -half_width, half_width, color=color, lw=3.0)
    axes[1].plot([-half_width, half_width], [y_pos, y_pos], marker="|", color=color, ms=12, lw=0.0)
    axes[1].text(half_width + 0.25, y_pos, rf"{label}: $\pm {half_width:.0f}\omega_0$", va="center", fontsize=9, color=color)
axes[1].axvline(0.0, color="0.2", lw=0.8)
axes[1].set_title("Bilateral bandwidth markers")
axes[1].set_xlabel(r"angular frequency / $\omega_0$")
axes[1].set_yticks([])
axes[1].set_xlim(-(max(bandwidth_over_omega0) + 1.5), max(bandwidth_over_omega0) + 1.5)
axes[1].set_ylim(min(row[3] for row in bandwidth_marker_rows) - 0.35, max(row[3] for row in bandwidth_marker_rows) + 0.45)
axes[1].grid(axis="x", alpha=0.3)
axes[1].set_xticks(np.arange(-max(bandwidth_over_omega0), max(bandwidth_over_omega0) + 1.0, 1.0))
plt.tight_layout()
plt.show()
```

## Stage-Wise ACF, Spectrum/PSD, and Bandwidth Analysis

This section makes the second-order analysis explicit for every stage of the assigned communication chain. The input is the finite zero-mean rectangular sequence

$$
m[n]=u[n]-u[n-2048]-\frac{1}{2},
\qquad
0\leq n<4096,
$$

with high level $1/2$, low level $-1/2$, duration $L=2048$ samples, and sampling increment $\Delta t=1/4096$.

### Input Stage

For the finite deterministic sequence, the finite-record autocorrelation is

$$
\widehat{R}_m[\ell]=\frac{1}{N-\ell}\sum_{n=0}^{N-1-\ell}(m[n]-\bar m)(m[n+\ell]-\bar m),
\qquad
\ell\geq 0.
$$

For the raw rectangular indicator, the nonzero-bin Fourier transform is the discrete Dirichlet-kernel form

$$
M(e^{j\Omega})=e^{-j\Omega(L-1)/2}\frac{\sin(\Omega L/2)}{\sin(\Omega/2)},
\qquad
L=2048,
$$

plus the finite-record interpretation implied by the sampled sequence. At the DFT bins $\Omega_k=2\pi k/N$, the theoretical normalized magnitude is

$$
\frac{|M[k]|}{N}
=\frac{1}{N}\left|\frac{\sin(\pi kL/N)}{\sin(\pi k/N)}\right|,
\qquad k\ne 0,
$$

For the actual centered message, the same expression holds at $k\ne 0$, but the constant offset removes the DC bin, so $M[0]=0$. Since $L=N/2$, the nonzero AC harmonics occur at odd $k$. This first stage is deterministic, so its spectrum should be computed directly from these DFT samples $M[k]$, not estimated as a PSD from data. The plotted spectrum is bilateral: the positive- and negative-frequency DFT samples are displayed together.

There are two different bandwidth references in this finite experiment. The pulse-duration reference uses $\tau=L\Delta t$ and the first continuous rectangular-pulse zero,

$$
B_{\omega,\tau}=\frac{2\pi}{\tau}.
$$

The simulation-duration reference uses the DFT frequency grid generated by $T=N\Delta t$, with spacing

$$
\omega_0=\frac{2\pi}{T}.
$$

For the default input-stage display and metric, the notebook keeps only the first three nonzero odd harmonics $k\in\{1,3,5\}$, corresponding to $f\in\{f_m,3f_m,5f_m\}$. The one-sided angular bandwidth on this finite simulation grid is therefore

$$
B_{\omega,\mathrm{in}}=5\omega_0.
$$

The notebook also compares several retained-harmonic sets, $\{1\}$, $\{1,3\}$, $\{1,3,5\}$, $\{1,3,5,7\}$, and $\{1,3,5,7,9\}$, so that the bandwidth scaling is visible rather than tested only at one truncation. A separate diagnostic varies the pair $(\tau,T)$ and displays, row by row, the time-domain pulse, the bilateral DFT samples, and the bandwidth markers. This emphasizes the duty cycle

$$
D=\frac{\tau}{T},
\qquad
\frac{B_{\omega,\tau}}{\omega_0}=\frac{T}{\tau}=\frac{1}{D}.
$$

The DC component is zero by construction, and higher sidelobes are intentionally excluded from the retained-harmonic input-stage summary. PSD estimation is reserved for the noisy channel and receiver-output stages.

### DSB-SC Modulation Stage

For

$$
s_{\mathrm{SC}}[n]=A_c m[n]\cos(2\pi f_c n\Delta t),
$$

the cycle-averaged PSD is a shifted copy of the input spectrum:

$$
\overline{S}_{s_{\mathrm{SC}}}(f)=\frac{A_c^2}{4}\left[S_m(f-f_c)+S_m(f+f_c)\right].
$$

The nominal sideband span around the carrier is determined by the practical baseband cutoff selected for the rectangular-pulse transition content.

### DSB-LC Modulation Stage

For

$$
s_{\mathrm{LC}}[n]=A_c[1+\mu m[n]]\cos(2\pi f_c n\Delta t),
$$

the PSD contains the carrier component plus sidebands generated by the rectangular pulse:

$$
\overline{S}_{s_{\mathrm{LC}}}(f)
=\frac{A_c^2}{4}\left[\delta(f-f_c)+\delta(f+f_c)\right]
+\frac{A_c^2\mu^2}{4}\left[S_m(f-f_c)+S_m(f+f_c)\right].
$$

Increasing $\mu$ increases the sideband power relative to the carrier. Since the assigned zero-mean pulse satisfies $m[n]\in\{1/2,-1/2\}$, the envelope multiplier $1+\mu m[n]$ does not become negative for the requested $\mu$ values.

### AWGN Channel Stage

The received signal is

$$
r[n]=s[n]+w[n],
$$

where $w[n]$ is white Gaussian noise independent of $s[n]$. Hence

$$
R_r[\ell]=R_s[\ell]+R_w[\ell],
\qquad
S_r(f)=S_s(f)+S_w(f).
$$

White noise spreads energy across the Nyquist band, so the empirical occupied bandwidth of the raw received waveform grows as SNR decreases.

### Envelope-Detector Stage

For LC with good inertia,

$$
\widehat{m}_{\mathrm{env}}[n]\approx m[n].
$$

For SC with this bipolar zero-mean rectangular pulse, envelope detection returns magnitude rather than the signed message in the noiseless ideal case. The bad-inertia detector has an overly slow discharge and therefore stretches the falling edge; analytically this behaves like nonlinear envelope extraction followed by excessive smoothing, which changes the ACF and narrows or distorts the useful baseband spectrum.

### Coherent-Demodulator Stage

For phase error $\phi$,

$$
\widehat{m}_{\mathrm{coh}}[n]\approx m[n]\cos\phi + n_b[n],
$$

after the correct LC DC removal or SC scaling. Therefore

$$
R_{\widehat{m}_{\mathrm{coh}}}[\ell]\approx \cos^2\phi\,R_m[\ell]+R_{n_b}[\ell],
\qquad
S_{\widehat{m}_{\mathrm{coh}}}(f)\approx \cos^2\phi\,S_m(f)+S_{n_b}(f).
$$

### IQ-Demodulator Stage

An ideal IQ receiver estimates the complex envelope. After phase correction,

$$
\widehat{u}[n]\approx A_c m[n]+n_I[n]+j n_Q[n]
$$

for SC, and similarly $A_c[1+\mu m[n]]$ for LC before DC removal. IQ imbalance mixes and rescales the two quadratures, creating residual image leakage and changing the measured PSD and ACF even when pure phase rotation has been corrected.


```python
def empirical_acf(samples, sample_rate_hz, max_lag_s=0.10, normalize=True):
    samples = np.asarray(samples, dtype=float)
    centered = samples - np.mean(samples)
    n = centered.size
    max_lag = min(int(max_lag_s * sample_rate_hz), n - 1)
    nfft = 1 << int(np.ceil(np.log2(2 * n - 1)))
    spectrum = np.fft.rfft(centered, n=nfft)
    acf = np.fft.irfft(np.abs(spectrum) ** 2, n=nfft)[: max_lag + 1]
    acf = acf / np.arange(n, n - max_lag - 1, -1)
    if normalize and abs(acf[0]) > 1e-15:
        acf = acf / acf[0]
    return np.arange(max_lag + 1) / sample_rate_hz, acf


def empirical_psd(samples, sample_rate_hz):
    samples = np.asarray(samples, dtype=float)
    centered = samples - np.mean(samples)
    window = np.hanning(centered.size)
    spectrum = np.fft.rfft(centered * window)
    psd = np.abs(spectrum) ** 2 / (sample_rate_hz * np.sum(window**2))
    if centered.size % 2 == 0:
        psd[1:-1] *= 2.0
    else:
        psd[1:] *= 2.0
    freqs = np.fft.rfftfreq(centered.size, d=1.0 / sample_rate_hz)
    return freqs, psd


def deterministic_fourier_spectrum(samples, sample_rate_hz, harmonic_bins=None):
    samples = np.asarray(samples, dtype=float)
    spectrum = np.fft.rfft(samples)
    spectral_energy = np.abs(spectrum) ** 2 / samples.size**2
    if samples.size % 2 == 0:
        spectral_energy[1:-1] *= 2.0
    else:
        spectral_energy[1:] *= 2.0
    freqs = np.fft.rfftfreq(samples.size, d=1.0 / sample_rate_hz)
    if harmonic_bins is not None:
        harmonic_bins = np.asarray(harmonic_bins, dtype=int)
        selected = np.zeros_like(spectral_energy)
        valid = harmonic_bins[(0 <= harmonic_bins) & (harmonic_bins < spectral_energy.size)]
        selected[valid] = spectral_energy[valid]
        spectral_energy = selected
    return freqs, spectral_energy


def occupied_bandwidth(freqs, spectral_values, energy_fraction=0.99, discrete=False):
    freqs = np.asarray(freqs, dtype=float)
    spectral_values = np.clip(np.asarray(spectral_values, dtype=float), 0.0, None)
    if spectral_values.size < 2 or np.sum(spectral_values) <= 0.0:
        return 0.0, 0.0, 0.0
    if discrete:
        cumulative = np.cumsum(spectral_values)
    else:
        cumulative = np.zeros_like(spectral_values)
        cumulative[1:] = np.cumsum(0.5 * (spectral_values[1:] + spectral_values[:-1]) * np.diff(freqs))
    total = cumulative[-1]
    if total <= 0.0:
        return 0.0, 0.0, 0.0
    lower_target = 0.5 * (1.0 - energy_fraction) * total
    upper_target = (1.0 - 0.5 * (1.0 - energy_fraction)) * total
    if discrete:
        f_low = float(freqs[min(np.searchsorted(cumulative, lower_target, side="left"), freqs.size - 1)])
        f_high = float(freqs[min(np.searchsorted(cumulative, upper_target, side="left"), freqs.size - 1)])
        return f_low, f_high, f_high - f_low
    f_low = float(np.interp(lower_target, cumulative, freqs))
    f_high = float(np.interp(upper_target, cumulative, freqs))
    return f_low, f_high, f_high - f_low


def stage_second_order_summary(stage_name, samples, sample_rate_hz, expected_band_hz=None, spectrum_method="psd_estimate", harmonic_bins=None):
    samples = np.asarray(samples, dtype=float)
    lag, acf = empirical_acf(samples, sample_rate_hz)
    if spectrum_method == "deterministic_fourier":
        freqs, spectral_values = deterministic_fourier_spectrum(samples, sample_rate_hz, harmonic_bins=harmonic_bins)
        power = float(np.mean(samples**2))
        spectrum_kind = "DFT"
        discrete_spectrum = True
        dominant_frequency = float(freqs[int(np.argmax(spectral_values))]) if spectral_values.size else 0.0
        active = np.flatnonzero(spectral_values > 0.0)
        f_low, f_high, bw_99 = (0.0, float(freqs[active[-1]]), float(freqs[active[-1]])) if active.size else (0.0, 0.0, 0.0)
    elif spectrum_method == "psd_estimate":
        freqs, spectral_values = empirical_psd(samples, sample_rate_hz)
        power = float(np.mean((samples - np.mean(samples)) ** 2))
        spectrum_kind = "PSD"
        discrete_spectrum = False
        dominant_frequency = float(freqs[int(np.argmax(spectral_values[1:])) + 1]) if spectral_values.size > 1 else 0.0
        f_low, f_high, bw_99 = occupied_bandwidth(freqs, spectral_values, energy_fraction=0.99, discrete=discrete_spectrum)
    else:
        raise ValueError(f"Unknown spectrum_method: {spectrum_method}")
    return {
        "stage": stage_name,
        "spectrum_kind": spectrum_kind,
        "power": power,
        "acf_1ms": float(np.interp(0.001, lag, acf)),
        "acf_10ms": float(np.interp(0.010, lag, acf)),
        "peak_freq_hz": dominant_frequency,
        "f_low_99": f_low,
        "f_high_99": f_high,
        "bw_99": bw_99,
        "expected_bw": expected_band_hz,
        "lag": lag,
        "acf": acf,
        "freqs": freqs,
        "psd": spectral_values,
    }


def compact_stage_row(summary):
    expected = "-" if summary["expected_bw"] is None else f"{summary['expected_bw']:.1f}"
    expected_omega = "-" if summary["expected_bw"] is None else f"{summary['expected_bw'] / input_signal_frequency_hz:.1f}"
    return {
        "stage": summary["stage"],
        "spectrum": summary["spectrum_kind"],
        "power": f"{summary['power']:.4f}",
        "acf_1ms": f"{summary['acf_1ms']:+.3f}",
        "acf_10ms": f"{summary['acf_10ms']:+.3f}",
        "peak_hz": f"{summary['peak_freq_hz']:.1f}",
        "f_low_99": f"{summary['f_low_99']:.1f}",
        "f_high_99": f"{summary['f_high_99']:.1f}",
        "bw_99": f"{summary['bw_99']:.1f}",
        "expected_bw": expected,
        "expected_B/omega0": expected_omega,
    }


def build_stage_signals_for_second_order(mode, mu, snr_db, phi, iq_case_name):
    transmitted = modulate(message, mode, mu)
    received = add_awgn(transmitted, snr_db, rng)
    good_env, _ = envelope_demodulate(received, mode, mu, envelope_rc_cases["good inertia"])
    bad_env, _ = envelope_demodulate(received, mode, mu, envelope_rc_cases["bad inertia"])
    coherent, _ = coherent_demodulate(received, mode, mu, phi)
    iq_params = iq_imbalance_cases[iq_case_name]
    iq_recovered, _ = iq_demodulate(
        received,
        mode,
        mu,
        phi,
        gain_imbalance=iq_params["gain"],
        phase_skew_rad=np.deg2rad(iq_params["phase_deg"]),
        correct_phase=True,
    )
    return {
        "input message": (message, message_bandwidth_hz),
        f"{mode} modulated": (transmitted, 2.0 * message_bandwidth_hz),
        f"{mode} + AWGN": (received, None),
        f"{mode} env good": (good_env, message_bandwidth_hz),
        f"{mode} env bad": (bad_env, None),
        f"{mode} coherent": (coherent, message_bandwidth_hz),
        f"{mode} IQ {iq_case_name}": (iq_recovered, message_bandwidth_hz),
    }


analysis_mu = 1.00
analysis_snr_db = 10.0
analysis_phi = np.pi / 4.0
analysis_iq_case = "imbalanced IQ"

stage_summaries_by_mode = {}
for mode in ["LC", "SC"]:
    stage_signals = build_stage_signals_for_second_order(
        mode=mode,
        mu=analysis_mu,
        snr_db=analysis_snr_db,
        phi=analysis_phi,
        iq_case_name=analysis_iq_case,
    )
    summaries = []
    for name, (values, expected_bw) in stage_signals.items():
        if name == "input message":
            summaries.append(
                stage_second_order_summary(
                    name,
                    values,
                    fs,
                    expected_band_hz=input_harmonic_bandwidth_hz,
                    spectrum_method="deterministic_fourier",
                    harmonic_bins=input_harmonic_bins,
                )
            )
        else:
            summaries.append(stage_second_order_summary(name, values, fs, expected_band_hz=expected_bw, spectrum_method="psd_estimate"))
    stage_summaries_by_mode[mode] = summaries

    print(f"\nSecond-order summary for {mode}: mu={analysis_mu:.2f}, SNR={analysis_snr_db:.0f} dB, phi=pi/4, IQ={analysis_iq_case}")
    print(
        format_table(
            [compact_stage_row(item) for item in summaries],
            ["stage", "spectrum", "power", "acf_1ms", "acf_10ms", "peak_hz", "f_low_99", "f_high_99", "bw_99", "expected_bw", "expected_B/omega0"],
        )
    )

# Compact bandwidth trends across the requested mu/SNR grid for major stages.
bandwidth_grid_rows = []
for mode in ["LC", "SC"]:
    for mu in modulation_indices:
        for snr_db in snr_db_values:
            signals = build_stage_signals_for_second_order(
                mode=mode,
                mu=mu,
                snr_db=snr_db,
                phi=np.pi / 4.0,
                iq_case_name="imbalanced IQ",
            )
            for stage_name in [f"{mode} modulated", f"{mode} + AWGN", f"{mode} env good", f"{mode} coherent", f"{mode} IQ imbalanced IQ"]:
                values, expected_bw = signals[stage_name]
                freqs, psd = empirical_psd(values, fs)
                f_low, f_high, bw_99 = occupied_bandwidth(freqs, psd)
                bandwidth_grid_rows.append(
                    {
                        "mode": mode,
                        "mu": f"{mu:.2f}",
                        "snr": f"{snr_db:.0f}",
                        "stage": stage_name.replace(f"{mode} ", ""),
                        "bw_99": bw_99,
                        "f_low": f_low,
                        "f_high": f_high,
                    }
                )

bandwidth_summary = []
for mode in ["LC", "SC"]:
    for stage in ["modulated", "+ AWGN", "env good", "coherent", "IQ imbalanced IQ"]:
        values = [row for row in bandwidth_grid_rows if row["mode"] == mode and row["stage"] == stage]
        bw_values = np.array([row["bw_99"] for row in values])
        bandwidth_summary.append(
            {
                "mode": mode,
                "stage": stage,
                "min_bw_99": f"{np.min(bw_values):.1f}",
                "median_bw_99": f"{np.median(bw_values):.1f}",
                "max_bw_99": f"{np.max(bw_values):.1f}",
            }
        )

print("\n99% occupied-bandwidth trend across all requested modulation indices and SNR values")
print(format_table(bandwidth_summary, ["mode", "stage", "min_bw_99", "median_bw_99", "max_bw_99"]))

# ACF and spectrum/PSD visual comparison for each stage in the diagnostic case.
fig, axes = plt.subplots(2, 2, figsize=(15, 9))
for col, mode in enumerate(["LC", "SC"]):
    for summary in stage_summaries_by_mode[mode]:
        axes[0, col].plot(summary["lag"] * 1e3, summary["acf"], lw=1.4, label=summary["stage"])
        positive = summary["freqs"] > 0.0
        label = f"{summary['stage']} ({summary['spectrum_kind']})"
        axes[1, col].semilogy(summary["freqs"][positive], summary["psd"][positive] + 1e-18, lw=1.2, label=label)
    axes[0, col].set_title(f"{mode}: normalized ACF by stage")
    axes[0, col].set_xlabel("lag [ms]")
    axes[0, col].set_ylabel("normalized ACF")
    axes[0, col].set_xlim(0.0, 60.0)
    axes[0, col].grid(alpha=0.3)
    axes[1, col].set_title(f"{mode}: DFT spectrum / PSD by stage")
    axes[1, col].set_xlabel("frequency [Hz]")
    axes[1, col].set_ylabel("DFT energy or PSD")
    axes[1, col].set_xlim(1.0, 200.0)
    axes[1, col].grid(alpha=0.3)
    axes[1, col].legend(fontsize=8)
plt.tight_layout()
plt.show()
```

### Empirical Interpretation

The tables and plots should be read together with the analytical receiver contracts.

- The input message is a zero-mean rectangular pulse with levels $+1/2$ and $-1/2$ and duration 2048 samples inside a 4096-sample record. Its sharp edges create sinc-like spectral sidelobes, and its DC bin is zero by construction.
- For LC with this zero-mean rectangular pulse, all requested modulation indices keep $1+\mu m[n]$ positive, so the envelope does not fold. Larger $\mu$ mainly increases envelope depth and sideband power.
- For SC with this bipolar pulse, envelope detection loses the sign in the ideal noiseless case; its practical error includes sign loss, noise, and detector inertia.
- Bad envelope inertia smooths too aggressively. It reduces carrier ripple, but it also stretches the falling edge of the rectangular pulse, which changes the ACF, PSD, and waveform error.
- Coherent demodulation degrades with $\phi$ according to the analytical factor $\cos\phi$. The shape can remain high after affine calibration, but the raw NMSE grows because the receiver calibrated for zero phase receives the wrong amplitude and, for LC, the wrong DC level.
- IQ demodulation can correct a pure carrier phase rotation by rotating the recovered complex envelope. IQ imbalance is different: unequal I/Q gains and quadrature skew deform the complex plane, so phase correction alone does not remove all error.


## 13. Final Conclusions

1. DSB-SC modulation translates the baseband spectrum to symmetric sidebands and suppresses the carrier line. Its transmitted amplitude changes sign with $m(t)$, so envelope detection recovers $|m(t)|$ rather than $m(t)$.

2. DSB-LC adds a carrier bias. When $1+k_a m(t)\geq 0$, the passband envelope is exactly proportional to $1+k_a m(t)$, so a simple envelope detector can recover the message after DC removal and gain correction.

3. Coherent demodulation multiplies by a synchronized sinusoid and low-pass filters. It preserves the signed message and, for WSS inputs, preserves the autocorrelation and PSD shape up to a scale factor.

4. A phase error $\phi$ in coherent demodulation introduces a factor $\cos\phi$ in amplitude and $\cos^2\phi$ in power. The special case $\phi=\pi/2$ gives a null output.

5. IQ demodulation is the general baseband method. It recovers the complex envelope $u(t)$, from which amplitude, phase, frequency offset, carrier phase, and signed DSB-SC information can be analyzed.

6. The correct receiver depends on the transmitted signal contract: envelope detection matches DSB-LC under no overmodulation, while DSB-SC requires coherent or IQ demodulation for signed message recovery.

7. In the assigned rectangular-pulse experiment, increasing AWGN lowers all receiver metrics, but the dominant impairment depends on receiver type: envelope detection is mainly limited by RC inertia and noise, coherent detection is mainly limited by the $\cos\phi$ phase-error factor, and IQ detection can remove pure phase rotation but remains sensitive to I/Q gain and quadrature imbalance.

8. The notebook now verifies both quadrature and canonical representations analytically and empirically: LC and SC are written as $I[n]\cos(2\pi f_c n\Delta t)-Q[n]\sin(2\pi f_c n\Delta t)$ and as $A[n]\cos(2\pi f_c n\Delta t+\varphi[n])$, while the IQ experiment estimates $\widehat{I}$, $\widehat{Q}$, $\widehat{A}$, and $\widehat{\varphi}$ under ideal and imbalanced IQ conditions.
