# 3.2. Narrowband Random Signals

A stationary random process $\xi(t)$ with zero mathematical expectation is called **narrowband** if the angular bandwidth $\Delta \omega$ of the frequency region where the spectral density $S_\xi(\omega)$ is practically nonzero is small compared to some central angular frequency $\omega_0$ of this region, i.e.,

$$
S_\xi(\omega) \neq 0 \quad \text{for} \quad \omega_0 - \Delta \omega / 2 < \omega < \omega_0 + \Delta \omega / 2, \qquad \Delta \omega \ll \omega_0,
\tag{3.6}
$$

where $S_\xi(\omega)$ denotes the one-sided power spectral density (per rad/s), and $\Delta \omega$ is the effective angular bandwidth, which can be defined in various ways (e.g., 3-dB bandwidth, RMS bandwidth, or equivalent noise bandwidth).

If the one-sided spectral density $S_\xi(\omega)$ is symmetric with respect to some angular frequency $\omega_0$, then it is natural to select this frequency as the central frequency. In other cases, $\omega_0$ may be chosen more or less arbitrarily. A common choice is the spectral centroid (i.e., the first moment of the normalized spectral density):

$$
\omega_0 = \left( \int_0^{\infty} S_\xi(\omega) \, d\omega \right)^{-1}
\int_0^{\infty} \omega S_\xi(\omega) \, d\omega.
\tag{3.7}
$$

The ACF of a narrowband random process can always be represented in the form [4]:

$$
R_\xi(\tau) = \sigma_\xi^2 r_c(\tau) \cos(\omega_0 \tau) - \sigma_\xi^2 r_s(\tau) \sin(\omega_0 \tau)
= \sigma_\xi^2 r(\tau) \cos\bigl(\omega_0 \tau + \gamma(\tau)\bigr),
\tag{3.8}
$$

where

$$
r(\tau) = \sqrt{r_c^2(\tau) + r_s^2(\tau)},
\qquad
\tan \gamma(\tau) = \frac{r_s(\tau)}{r_c(\tau)},
\tag{3.9a}
$$

$$
\sigma_\xi^2 r_c(\tau) = \frac{1}{2\pi} \int_{-\omega_0}^{\infty} S_\xi(\omega_0 + \nu) \cos(\nu \tau) \, d\nu,
\tag{3.9b}
$$

$$
\sigma_\xi^2 r_s(\tau) = \frac{1}{2\pi} \int_{-\omega_0}^{\infty} S_\xi(\omega_0 + \nu) \sin(\nu \tau) \, d\nu,
\tag{3.9c}
$$

with the initial conditions

$$
r_c(0) = 1, \qquad r_s(0) = 0, \qquad \gamma(0) = 0.
\tag{3.10}
$$

Here, $\sigma_\xi^2 = R_\xi(0)$ denotes the variance of the narrowband process $\xi(t)$.

The functions $r_c(\tau)$ and $r_s(\tau)$ are respectively the in-phase and quadrature correlation components. The envelope $r(\tau)$ and phase modulation $\gamma(\tau)$ characterize the slow variations of the narrowband process relative to the carrier frequency $\omega_0$.

In the particular case where the spectral density $S_\xi(\omega)$ is symmetric with respect to the central angular frequency $\omega_0$, the ACF of the narrowband process simplifies to

$$
R_\xi(\tau) = \sigma_\xi^2 r(\tau) \cos(\omega_0 \tau)
\simeq \frac{\cos(\omega_0 \tau)}{2\pi} \int_{-\infty}^{\infty} S_\xi(\omega_0 + \nu) \cos(\nu \tau) \, d\nu,
\tag{3.11}
$$

since

$$
r_s(\tau) = 0, \qquad \gamma(\tau) = 0.
\tag{3.12}
$$

Because the right-hand sides of Equations (3.9b) and (3.9c) involve the spectral density of the process $\xi(t)$ shifted to baseband (i.e., the low-frequency spectrum), the functions $r_c(\tau)$, $r_s(\tau)$, as well as $r(\tau)$ and $\gamma(\tau)$, vary slowly compared to $\cos(\omega_0 \tau)$.

Sample functions (realizations) of a narrowband process $\xi(t)$ exhibit the structure of a modulated harmonic oscillation. Consequently, narrowband processes are commonly represented in the canonical form

$$
\xi(t) = A(t) \cos\bigl(\omega_0 t + \varphi(t)\bigr),
\qquad
A(t) \ge 0,
\qquad
-\pi < \varphi(t) \le \pi,
\tag{3.13}
$$

where $A(t)$ and $\varphi(t)$ are slowly varying functions compared to $\cos(\omega_0 t)$, termed the envelope and phase of the narrowband process $\xi(t)$, respectively.

It is also useful to introduce the concept of instantaneous angular frequency, defined as

$$
\omega(t) = \omega_0 + \dot{\varphi}(t),
\tag{3.14}
$$

where the overdot denotes differentiation with respect to time.

The representation of the narrowband process in Equation (3.13) can alternatively be expressed as

$$
\xi(t) = A_c(t) \cos(\omega_0 t) - A_s(t) \sin(\omega_0 t),
\tag{3.15}
$$

where

$$
A_c(t) = A(t) \cos \varphi(t),
\qquad
A_s(t) = A(t) \sin \varphi(t).
\tag{3.16}
$$

Consequently, it follows that

$$
A(t) = \sqrt{A_c^2(t) + A_s^2(t)},
\qquad
\varphi(t) = \arctan\!\left(\frac{A_s(t)}{A_c(t)}\right),
\tag{3.17}
$$

$$
\omega(t) = \omega_0 + \frac{\dot{A}_s(t) A_c(t) - \dot{A}_c(t) A_s(t)}{A_c^2(t) + A_s^2(t)}.
\tag{3.18}
$$

The representations in Equations (3.13) and (3.15) admit a geometric interpretation: the envelope $A(t)$ is the magnitude of a vector whose projections onto the axes of a rectangular coordinate system are $A_c(t)$ and $A_s(t)$, and the phase $\varphi(t)$ is the angle subtended between the abscissa axis and the vector direction. Both $A(t)$ and $\varphi(t)$ vary randomly in time, causing the tip of the vector to execute a random walk in the plane; the narrowband process $\xi(t)$ can thus be visualized as a rotating vector with randomly fluctuating amplitude and phase. The projections $A_c(t)$ and $A_s(t)$ are accordingly termed the quadrature components of the process $\xi(t)$.

# Task

Show that if $\xi(t)$ is ruled by a Gaussian distribution, then

- $A_c(t)$ and $A_s(t)$ are also Gaussian-distributed, each one having variance $\sigma_c^2 = \sigma_s^2 = \sigma_\xi^2$.
- $A_c(t)$ and $A_s(t)$ are zero-mean, if $\xi(t)$ is zero mean.
- $A_c(t)$ and $A_s(t)$ are zero-correlated regardless of the lag $\tau$, if $S_\xi(\omega)$ is symmetric with respect to $\omega_0$.

## Worked Solution

A standard way to obtain the quadrature components is by coherent demodulation followed by ideal low-pass filtering:

$$
A_c(t) = \operatorname{LPF}\!\left\{ 2\,\xi(t)\cos(\omega_0 t) \right\},
\qquad
A_s(t) = \operatorname{LPF}\!\left\{ -2\,\xi(t)\sin(\omega_0 t) \right\}.
$$

These operations are linear: multiplication by a deterministic sinusoid is linear, and ideal low-pass filtering is also linear. Therefore, if $\xi(t)$ is Gaussian, then $A_c(t)$ and $A_s(t)$ are Gaussian as well, because linear transformations preserve Gaussianity.

To determine their second-order properties, define

$$
R_{cc}(\tau) = \mathbb{E}\{A_c(t)A_c(t+\tau)\},
\qquad
R_{ss}(\tau) = \mathbb{E}\{A_s(t)A_s(t+\tau)\},
$$

$$
R_{cs}(\tau) = \mathbb{E}\{A_c(t)A_s(t+\tau)\},
\qquad
R_{sc}(\tau) = \mathbb{E}\{A_s(t)A_c(t+\tau)\}.
$$

Using Equation (3.15),

$$
\xi(t) = A_c(t)\cos(\omega_0 t) - A_s(t)\sin(\omega_0 t).
$$

### 1. Gaussianity and Variance of $A_c(t)$ and $A_s(t)$

Expanding the autocorrelation of $\xi(t)$ gives

$$
R_\xi(\tau) = \mathbb{E}\{\xi(t)\xi(t+\tau)\}
$$

$$
= \frac{R_{cc}(\tau) + R_{ss}(\tau)}{2}\cos(\omega_0\tau)
+ \frac{R_{cc}(\tau) - R_{ss}(\tau)}{2}\cos(2\omega_0 t + \omega_0\tau)
$$

$$
- \frac{R_{cs}(\tau) + R_{sc}(\tau)}{2}\sin(2\omega_0 t + \omega_0\tau)
- \frac{R_{cs}(\tau) - R_{sc}(\tau)}{2}\sin(\omega_0\tau).
$$

Since $\xi(t)$ is stationary, $R_\xi(\tau)$ cannot depend on the absolute time $t$. Hence the coefficients of the terms containing $2\omega_0 t$ must vanish:

$$
R_{cc}(\tau) = R_{ss}(\tau),
\qquad
R_{sc}(\tau) = -R_{cs}(\tau).
$$

Therefore,

$$
R_\xi(\tau) = R_{cc}(\tau)\cos(\omega_0\tau) - R_{cs}(\tau)\sin(\omega_0\tau).
$$

Comparing this with Equation (3.8),

$$
R_\xi(\tau) = \sigma_\xi^2 r_c(\tau)\cos(\omega_0\tau) - \sigma_\xi^2 r_s(\tau)\sin(\omega_0\tau),
$$

we identify

$$
R_{cc}(\tau) = R_{ss}(\tau) = \sigma_\xi^2 r_c(\tau),
\qquad
R_{cs}(\tau) = \sigma_\xi^2 r_s(\tau).
$$

Now evaluate at $\tau = 0$. Since Equation (3.10) gives $r_c(0)=1$,

$$
\sigma_c^2 = R_{cc}(0) = \sigma_\xi^2 r_c(0) = \sigma_\xi^2,
$$

$$
\sigma_s^2 = R_{ss}(0) = \sigma_\xi^2 r_c(0) = \sigma_\xi^2.
$$

So both quadrature components are Gaussian-distributed and each has variance equal to the variance of $\xi(t)$.

### 2. Zero-Mean Property

Because $A_c(t)$ and $A_s(t)$ are obtained from $\xi(t)$ by linear operations, their means are linear functions of the mean of $\xi(t)$. If $\xi(t)$ is zero mean, then

$$
\mathbb{E}\{A_c(t)\} = 0,
\qquad
\mathbb{E}\{A_s(t)\} = 0.
$$

The same conclusion follows directly from Equation (3.15):

$$
\mathbb{E}\{\xi(t)\} = \mathbb{E}\{A_c(t)\}\cos(\omega_0 t) - \mathbb{E}\{A_s(t)\}\sin(\omega_0 t).
$$

If $\mathbb{E}\{\xi(t)\}=0$ for every $t$, the only way this identity can hold for all $t$ is

$$
\mathbb{E}\{A_c(t)\} = 0,
\qquad
\mathbb{E}\{A_s(t)\} = 0.
$$

Hence both quadrature components are zero mean whenever the original narrowband Gaussian process is zero mean.

### 3. Zero Correlation When $S_\xi(\omega)$ Is Symmetric About $\omega_0$

From Equation (3.9c),

$$
\sigma_\xi^2 r_s(\tau) = \frac{1}{2\pi}\int_{-\omega_0}^{\infty} S_\xi(\omega_0 + \nu)\sin(\nu\tau)\,d\nu.
$$

If the spectrum is symmetric with respect to $\omega_0$, then the shifted spectrum $S_\xi(\omega_0+\nu)$ is an even function of $\nu$, while $\sin(\nu\tau)$ is an odd function of $\nu$. Their product is odd, so the integral cancels. This is exactly the special case stated in Equation (3.12):

$$
r_s(\tau) = 0.
$$

Using the identification found above,

$$
R_{cs}(\tau) = \sigma_\xi^2 r_s(\tau) = 0 \qquad \text{for all } \tau.
$$

Therefore $A_c(t)$ and $A_s(t)$ are uncorrelated for every lag whenever the narrowband spectrum is symmetric around the carrier frequency $\omega_0$.

## Final Conclusions

If $\xi(t)$ is a narrowband Gaussian random process, then:

1. $A_c(t)$ and $A_s(t)$ are Gaussian because they are obtained from $\xi(t)$ through linear operations.
2. Their variances are

$$
\sigma_c^2 = \sigma_s^2 = \sigma_\xi^2.
$$

3. If $\xi(t)$ is zero mean, then

$$
\mathbb{E}\{A_c(t)\} = \mathbb{E}\{A_s(t)\} = 0.
$$

4. If $S_\xi(\omega)$ is symmetric about $\omega_0$, then

$$
R_{cs}(\tau) = 0 \qquad \forall\tau,
$$

so the two quadrature components are zero-correlated for all lags.
